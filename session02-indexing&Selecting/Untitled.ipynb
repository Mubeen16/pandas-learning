```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nba.csv")
df.head()
```

```python
# Step 2: Basic Grouping – Total Salary by Team
salary_by_team = df.groupby('Team')['Salary'].sum()
salary_by_team.sort_values(ascending=False)
```

```python
group_by_position = df.groupby("Position")["Salary"].mean()
```

```python
group_by_position
```

```python
# Step 3: Count of Players Per Team
player_per_team = df['Team'].value_counts()
player_per_team
```

```python
#Step 4: Average Age per Position
df.groupby("Position")["Age"].mean()
```

```python
#  Step 5: Plotting – Total Salary by Team
salary_by_team.sort_values(ascending=False).plot(kind="bar", figsize=(12, 6), title="Total Salary by NBA Team")
plt.ylabel("Total Salary ($)")
plt.show()
```

```python
#  Step 6: Plot – Players Per Team (Horizontal Bar)

players_per_team.sort_values().plot(kind="barh", figsize=(10, 8), title="Players per Team")
plt.xlabel("Number of Players")
plt.show()
```
